```python
from libs import get_data
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = get_data.get_data("ENPH")
```

```python
df.head()
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
h=df.plot('Date', 'Close')
```

```python
train_data = df.iloc[:1400]
```

```python
T = 3
S=train_data['Close'][1399]
```

```python
close_stats = train_data['Close'].describe()
mu= close_stats['mean']
vol = close_stats['std']*math.sqrt(T)
```

```python
vol
```

```python
daily_closes = np.random.normal((mu/T), vol/math.sqrt(T),T)+1

price_list=[S]

for x in daily_closes:
    price_list.append(price_list[-1]*x)
plt.plot(price_list)
plt.show()
print(price_list[-1])
```

```python
results = []
ax=plt.plot()
for i in range(2000):
    daily_closes = np.random.normal((mu/T), vol/math.sqrt(T),T)+1

    price_list=[S]

    for x in daily_closes:
        price_list.append(price_list[-1]*x)
    plt.plot(price_list)
    results.append(price_list[-1])
    
# plt.ylim(0,800)
plt.show()
```

```python
print(np.mean(results))
```
