```python
train_data = df.iloc[:1200]
```

```python
mu=train_data['Close'].describe()['mean']
vol=train_data['Close'].describe()['std']
```

```python
T = 252
S=train_data['Close'][1199]
```

```python
S
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
daily_closes = np.random.normal((mu/T), vol/math.sqrt(T),T)+1

price_list=[S]

for x in daily_closes:
    price_list.append(price_list[-1]*x)
plt.plot(price_list)
plt.show()
```

```python
results = []
for i in range(5000):
    daily_closes = np.random.normal((mu/T), vol/math.sqrt(T),T)+1

    price_list=[S]

    for x in daily_closes:
        price_list.append(price_list[-1]*x)
    plt.plot(price_list)
    results.append(price_list[-1])
    
plt.show()
```

```python
print(np.mean(results))
```

```python
np.percentile(results,5)
```

```python
np.percentile(results,95)
```

```python
plt.hist(results, bins=10)
plt.show()
```
